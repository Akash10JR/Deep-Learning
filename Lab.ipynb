 1.Construct and verify the performance of bi-level and multi-level confusion
matrix using Python.

```python
from sklearn.metrics import confusion_matrix
y_true_bi = [0,1,0,1,0,1,1,0,1,0]
y_pred_bi = [0,1,0,0,1,1,1,0,1,0]
cm_bi = confusion_matrix(y_true_bi, y_pred_bi)
print("Bi-Level Confusion Matrix:")
print(cm_bi)
y_true_multi = [0,1,2,0,1,2,0,1,2,0]
y_pred_multi = [0,2,1,0,1,2,0,0,2,1]
cm_multi = confusion_matrix(y_true_multi, y_pred_multi)
print("\nMulti-Level Confusion Matrix:")
print(cm_multi)
```

2.Verification of the overfitting issue by using a chosen numerical database using Python.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
print("Training Accuracy:", model.score(X_train, y_train))
print("Testing Accuracy:", model.score(X_test, y_test))
```

5.Demonstrating the performance of gradient descent algorithm using Python

```python
import numpy as np

# Sample Data
x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)

# Initial values
m = 0
c = 0
lr = 0.01      # learning rate
epochs = 1000  # number of iterations

# Gradient Descent
for i in range(epochs):
    y_pred = m*x + c
    error = y - y_pred

    m_grad = (-2/len(x)) * sum(x * error)
    c_grad = (-2/len(x)) * sum(error)

    m = m - lr * m_grad
    c = c - lr * c_grad

print("Slope (m):", m)
print("Intercept (c):", c)
```

6.Verifying the performance ofa image pre-processing with Otsu's function using
Python

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

img = data.coins()
img = img.astype(np.uint8)

thresh_val, otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

print("Optimal Otsu Threshold Value:", thresh_val)

plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title("Original")
plt.subplot(1,2,2); plt.imshow(otsu, cmap='gray'); plt.title("Otsu Result")
plt.show()
```

7.Verifying the performance of a image pre-processing with watershed algorithm
using Python.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

img = data.coins()
gray = img.astype(np.uint8)

kernel = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel, iterations=2)
sure_bg = cv2.dilate(opening, kernel, iterations=3)
dist = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
_, sure_fg = cv2.threshold(dist, 0.4 * dist.max(), 255, 0)
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg, sure_fg)
_, markers = cv2.connectedComponents(sure_fg)
markers = markers + 1
markers[unknown == 255] = 0
color_img = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)
markers = cv2.watershed(color_img, markers)
color_img[markers == -1] = [255, 0, 0]

plt.imshow(color_img)
plt.title("Watershed Segmentation")
plt.axis('off')
plt.show()
```

10.Performance evaluation of Linear regression Classifier using Breast cancer or
Iris database

```python
from sklearn import datasets, model_selection, metrics
from sklearn.linear_model import LinearRegression
import numpy as np

X, y = datasets.load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)
pred = np.rint(model.predict(X_test)).astype(int)

print("Accuracy:", metrics.accuracy_score(y_test, pred))
```

11.Performance evaluation of Logistic regression Classifier using Breast cancer or

12.Performance evaluation of Random Forest Classifier using Breast cancer or Iris
database

```python
from sklearn import datasets, model_selection, metrics
from sklearn.ensemble import RandomForestClassifier

X, y = datasets.load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("Accuracy:", metrics.accuracy_score(y_test, pred))
```

13. Performance evaluation of Decision Tree Classifier using Breast cancer or Iris
