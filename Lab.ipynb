 1.Construct and verify the performance of bi-level and multi-level confusion
matrix using Python.

```python
from sklearn.metrics import confusion_matrix
y_true_bi = [0,1,0,1,0,1,1,0,1,0]
y_pred_bi = [0,1,0,0,1,1,1,0,1,0]
cm_bi = confusion_matrix(y_true_bi, y_pred_bi)
print("Bi-Level Confusion Matrix:")
print(cm_bi)
y_true_multi = [0,1,2,0,1,2,0,1,2,0]
y_pred_multi = [0,2,1,0,1,2,0,0,2,1]
cm_multi = confusion_matrix(y_true_multi, y_pred_multi)
print("\nMulti-Level Confusion Matrix:")
print(cm_multi)
```

2.Verification of the overfitting issue by using a chosen numerical database using Python.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
print("Training Accuracy:", model.score(X_train, y_train))
print("Testing Accuracy:", model.score(X_test, y_test))
```

3.Analysing the performance of linear regression process using a chosen
numerical database using Python.

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X, y = load_diabetes(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)
model = LinearRegression()
model.fit(X_train, y_train)
print("Training Score:", model.score(X_train, y_train))
print("Testing Score:", model.score(X_test, y_test))
```

4.Analysing the performance of logistic regression process using a chosen
numerical database using Python.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
print("Training Accuracy:", model.score(X_train, y_train))
print("Testing Accuracy:", model.score(X_test, y_test))
```

5.Demonstrating the performance of gradient descent algorithm using Python

```python
import numpy as np

# Sample Data
x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)

# Initial values
m = 0
c = 0
lr = 0.01      # learning rate
epochs = 1000  # number of iterations

# Gradient Descent
for i in range(epochs):
    y_pred = m*x + c
    error = y - y_pred

    m_grad = (-2/len(x)) * sum(x * error)
    c_grad = (-2/len(x)) * sum(error)

    m = m - lr * m_grad
    c = c - lr * c_grad

print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
import cv2
import numpy as np

# Read image and convert to grayscale
img = cv2.imread('cells.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Threshold and remove noise
_, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
kernel = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)

# Apply watershed
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
_, sure_fg = cv2.threshold(dist_transform, 0.7*dist_transform.max(), 255, 0)
markers = cv2.connectedComponents(sure_fg.astype(np.uint8))[1]
cv2.watershed(img, markers)
img[markers == -1] = [255,0,0]  # Mark boundaries in red

cv2.imshow('Watershed Result', img)
cv2.waitKey(0)
```
